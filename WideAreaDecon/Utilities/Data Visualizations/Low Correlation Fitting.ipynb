```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import xlsxwriter
import seaborn as sns
from scipy import stats
import math
```

```python
# read in the entire dataset
master = pd.read_excel('..\Data Cleanup\cleaned_data.xlsx', sheet_name = 'master')

master = master[(master['EffMeas'] == 'LR') & (master['TotalApp'] == 1)]
```

```python
dFumigation = master[(master['AppMethod'] == 'fumigation') & (master['IndoorNonCarpet'] == 0) & (master['HVAC'] == 0) & 
                     (master['Roofing'] == 0) & (master['UndergroundNonCarpet'] == 0)]

print(dFumigation.Eff.count())

if dFumigation.Eff.count() > 0:
    binSize = round(1 + (3.322 * math.log10(dFumigation.Eff.count())))

    plt.hist(x = dFumigation.Eff.dropna(), bins=binSize, color='skyblue', edgecolor='black')

    plt.xlabel('Eff')
    plt.ylabel('Frequency')
    plt.title('Fumigation')

    plt.show()
```

```python
fumigationSurfaces = ['IndoorInterior', 'IndoorExterior', 'IndoorCarpet', 'IndoorCeilings', 'IndoorMisc', 'OutdoorExterior', 
                      'Pavement', 'Soil', 'Water', 'OutdoorMisc', 'UndergroundInterior', 'UndergroundCarpet', 
                      'UndergroundCeilings', 'UndergroundMisc']
```

```python
for sur in fumigationSurfaces:
    
    temp = dFumigation[(dFumigation[sur] == 1)]
    
    print(temp.Eff.count())

    if temp.Eff.count() > 0:
        binSize = round(1 + (3.322 * math.log10(temp.Eff.count())))

        plt.hist(x = temp.Eff.dropna(), bins=binSize, color='skyblue', edgecolor='black')

        plt.xlabel('Eff')
        plt.ylabel('Frequency')
        plt.title('Fumigation and ' + sur)

        plt.show()
```
